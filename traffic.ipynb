```python
# Cell 3: The Python script (run this last)

import cv2
from ultralytics import YOLO

# We need to install this here if it's a new session
!pip install -q ultralytics

print("\nüß† Loading the YOLO model...")
model = YOLO('yolov8m.pt')

input_video_path = 'road_traffic.mp4'
output_video_path = 'output_traffic.mp4'

print(f"üé• Attempting to open video file: {input_video_path}")
cap = cv2.VideoCapture(input_video_path)

if not cap.isOpened():
    print("‚ùå FATAL ERROR: cv2.VideoCapture() could not open the video file.")
else:
    print("‚úÖ Video file opened successfully. Proceeding to process frames...")
    w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    print(f"   Video properties: {w}x{h} at {fps:.2f} FPS")

    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_video_path, fourcc, fps, (w, h))

    print("\nüî• Starting video processing and object tracking...")
    frame_count = 0
    while True:
        success, frame = cap.read()
        if not success:
            print(f"\n- Could not read a new frame. End of video or error at frame {frame_count}.")
            break

        frame_count += 1
        print(f"--- Processing Frame {frame_count} ---")

        results = model.track(frame, persist=True)

        if results[0].boxes.id is not None:
            boxes = results[0].boxes.xywh.cpu()
            track_ids = results[0].boxes.id.int().cpu().tolist()
            class_ids = results[0].boxes.cls.int().cpu().tolist()
            class_names = results[0].names

            for box, track_id, cls_id in zip(boxes, track_ids, class_ids):
                x, y, w, h = box
                class_name = class_names[cls_id]
                print(f"  Object ID: {track_id}, Class: {class_name}, Position: (x={int(x)}, y={int(y)})")

        annotated_frame = results[0].plot()
        out.write(annotated_frame)

    cap.release()
    out.release()
    print(f"\n‚úÖ Video processing complete. Output saved to {output_traffic.mp4}")
```
